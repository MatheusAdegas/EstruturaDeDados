```python
global estoque_comida, estoque_bebida, i
estoque_comida = {
  "sanduiche":5,
  "bolo":6,
}
estoque_bebida = {
  "refrigerante": 10,
  "suco": 8,
  "cafe": 15,
}

def menu():
  mostrarOpcoes()
  i = input("Insira uma opção: ")
  while i != "5":
    if i == "0":
      repor_automatico()
    elif i == "1":
      mostrar_estoque()
    elif i == "2":
      adicionar_produto()
    elif i == "3":
      remover_produto()
    elif i == "4":
      consultar_produto()
    mostrarOpcoes()
    i = input("Insira uma opção: ")
  salvar_relatorio()

def mostrar_estoque():
  e = input("escolha: [b] bebida; [c] comida: ")
  print("\n")
  if e.lower() == "b":
    for i in estoque_bebida:
      print(f"{i}: {estoque_bebida[i]}")
  elif e.lower() == "c":
    for i in estoque_comida:
      print(f"{i}: {estoque_comida[i]}")
  else:
    print("estoque inválido.")
  print("\n")


def adicionar_produto():
  e = input("escolha: [b] bebida; [c] comida: ")
  x = input("Insira o nome do novo produto: ")
  y = input("Insira a quantidade do produto: ")
  if e.lower() == "b":
    if y.isdigit():
      estoque_bebida[x] = int(y)
      print("Produto adicionado com sucesso!")
    else:
      print("Quantidade inválida")
  elif e.lower() == "c":
    if y.isdigit():
      estoque_comida[x] = int(y)
      print("Produto adicionado com sucesso!")
    else:
      print("Quantidade inválida")
  else:
    print("estoque inválido.")

def remover_produto():
  e = input("escolha: [b] bebida; [c] comida: ")
  x = input("Insira o nome do produto afetado: ")
  y = input("Insira a quantidade a ser removida: ")
  while not y.isdigit():
    y = input("A quantidade deve ser um numero: ")
  if e.lower() == "b":
    if x in estoque_bebida:
      if int(y) < estoque_bebida[x]:
        estoque_bebida[x] -= int(y)
        print("Quantidade atualizada")
      else:
        estoque_bebida.pop(x)
        print("Produto removido com sucesso!")
  elif e.lower() == "c":
    if x in estoque_comida:
      if int(y) < estoque_comida[x]:
        estoque_comida[x] -= int(y)
        print("Quantidade atualizada")
      else:
        estoque_comida.pop(x)
        print("Produto removido com sucesso!")

def consultar_produto():
  e = input("escolha: [b] bebida; [c] comida: ")
  x = input("Insira o nome do produto a ser consultado: ")
  if e.lower() == "b":
    print(estoque_bebida[x])
  elif e.lower() == "c":
    print(estoque_comida[x])
  else:
    print("estoque inválido.")

def mostrarOpcoes():
  print("[0]: Repor automatico\n[1]: Mostrar estoque\n[2]: Adicionar produto\n[3]: Remover produto\n[4]: Consultar produto\n[5]: Sair e salvar")

def repor_automatico():
  for i in estoque_bebida:
    if estoque_bebida[i] < 3:
      estoque_bebida[i] += 5
  for i in estoque_comida:
    if estoque_comida[i] < 3:
      estoque_comida[i] += 5
  print("Estoque reposto.")

def salvar_relatorio():
  with open("estoque.txt", "w") as f:
    print("\nEstoque bebida: ")
    for x in estoque_bebida:
      print(f"{x}: {estoque_bebida[x]}")
    print("\nEstoque comida: ")
    for x in estoque_comida:
      print(f"{x}: {estoque_comida[x]}")

menu()
```
